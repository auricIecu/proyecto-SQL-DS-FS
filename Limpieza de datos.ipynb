```python
# Definir los mapeos para cada categoría
promocion_map = {
    'Septiembre': 1,
    'Noviembre': 2,
    'Febrero': 2,
    'Mayo': 4
}

campus_map = {
    'Madrid': 1,
    'Valencia': 2,
    'Vitoria': 3,
    'Bilbao': 4
}

modalidad_map = {
    'Online': 1,
    'Presencial': 2
}

vertical_map = {
    'FS': 1,
    'DS': 2,
    'CS': 3
}

# Generar las combinaciones en el orden del DataFrame
combinaciones = []
for _, row in alumnos_combinados_total_tabla.iterrows():
    combinacion = f"({promocion_map.get(row['Promocion'], 0)}," \
                  f"{campus_map.get(row['Campus'], 0)}," \
                  f"{modalidad_map.get(row['Modalidad'], 0)}," \
                  f"{vertical_map.get(row['Vertical'], 0)})"
    combinaciones.append(combinacion)
    print(combinacion)  # Imprime cada combinación en una línea para facilitar copia/pega
```
